```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble  import RandomForestClassifier
import sklearn.cross_decomposition
%matplotlib inline
```

```python
data=pd.read_csv("mnist_test.csv")
```

```python
data.head()
```

```python
a=data.iloc[3,1:].values
a=a.reshape(28,28).astype('uint8')
plt.imshow(a)
```

```python
#preparing the data
df_x=data.iloc[:,1:]
df_y=data.iloc[:,0]
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
#check data
y_train.head()
```

```python
# call rf classifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
pred=rf.predict(x_test)
pred
```

```python
#check prediction accuracy
a=y_test.values
count=0
for i in range(len(pred)):
    if pred[i]==a[i]:
        count=count+1
count
```
